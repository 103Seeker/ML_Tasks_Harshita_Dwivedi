```python
# Scenario-Based Teaching: Smart Travel Agent
# 🧑‍🏫 How You Start in Class
# Say this:
# “Imagine you are very busy and you tell your assistant:
# ‘Find me the best flight from Delhi to Mumbai.’
# Now instead of replying with just text, your assistant actually thinks,
# checks data, and gives a final decision.”

# Step 1 — Define Tools.👉 “Tools = functions the agent can call”

def search_flights(source, destination):
    return [
        {"airline": "IndiGo", "price": 4500, "time": "6:00 AM"},
        {"airline": "Air India", "price": 5200, "time": "9:00 AM"}
    ]

def check_weather(city):
    return {"city": city, "temperature": 28, "condition": "Clear"}

def calculate_budget(flights):
    cheapest = min(flights, key=lambda x: x["price"])
    return cheapest
# Step 2 — Agent Brain (Decision + Planning)
def agent_brain(user_query):
    plan = []

    if "flight" in user_query.lower():
        plan.append("search_flights")
        plan.append("check_weather")
        plan.append("choose_best")

    return plan
# Step 3 — Execution Engine
def execute_plan(plan):
    context = {}

    for step in plan:
        if step == "search_flights":
            print("\n[Action] Searching Flights...")
            context["flights"] = search_flights("Delhi", "Mumbai")
            print(context["flights"])

        elif step == "check_weather":
            print("\n[Action] Checking Weather...")
            context["weather"] = check_weather("Mumbai")
            print(context["weather"])

        elif step == "choose_best":
            print("\n[Action] Choosing Best Option...")
            best = calculate_budget(context["flights"])
            context["best_flight"] = best
            print(best)

    return context
  # Step 4 — Full Agent
def travel_agent(user_query):
    print("User Query:", user_query)

    # Step 1: Plan
    plan = agent_brain(user_query)
    print("\n[Plan]", plan)

    # Step 2: Execute
    result = execute_plan(plan)

    # Step 3: Final Response
    final_output = f"""
Best Flight: {result['best_flight']['airline']} at ₹{result['best_flight']['price']}
Weather: {result['weather']['condition']}, {result['weather']['temperature']}°C
Recommendation: Book the cheapest early flight.
"""
    return final_output
# Step 5 — Run It
response = travel_agent("Find best flight from Delhi to Mumbai")
print("\nFinal Answer:\n", response)
```

```python
# Scenario: Smart Shopping Assistant 🛒
# You say to your students:
# “Imagine you are planning a party and you tell your assistant:
# ‘Find me the best chocolate cake under ₹500.’
# Now instead of just replying with text, your assistant actually thinks, checks different bakery websites, compares prices, and then gives
# you the final decision.”
def search_cakes(product, max_price):
    return [
        {"name": "Chocolate Truffle Cake", "price": 450, "shop": "Sweet Treats"},
        {"name": "Dark Chocolate Cake", "price": 550, "shop": "Cake World"},
        {"name": "Chocolate Cream Cake", "price": 400, "shop": "Bake House"}
    ]

def check_ratings(shop):
    ratings = {
        "Sweet Treats": 4.5,
        "Cake World": 4.2,
        "Bake House": 4.7
    }
    return {"shop": shop, "rating": ratings.get(shop, 4.0)}

def calculate_best_option(cakes, max_price):
    # Filter under budget
    affordable = [cake for cake in cakes if cake["price"] <= max_price]

    # Pick cheapest
    best = min(affordable, key=lambda x: x["price"])
    return best
def agent_brain(user_query):
    plan = []

    if "cake" in user_query.lower():
        plan.append("search_cakes")
        plan.append("check_ratings")
        plan.append("choose_best")

    return plan
def execute_plan(plan):
    context = {}

    for step in plan:
        if step == "search_cakes":
            print("\n[Action] Searching Cakes...")
            context["cakes"] = search_cakes("chocolate cake", 500)
            print(context["cakes"])

        elif step == "check_ratings":
            print("\n[Action] Checking Ratings...")
            ratings = []
            for cake in context["cakes"]:
                rating_info = check_ratings(cake["shop"])
                ratings.append(rating_info)
            context["ratings"] = ratings
            print(ratings)

        elif step == "choose_best":
            print("\n[Action] Choosing Best Option...")
            best = calculate_best_option(context["cakes"], 500)
            context["best_cake"] = best
            print(best)

    return context
def shopping_agent(user_query):
    print("User Query:", user_query)

    # Step 1: Plan
    plan = agent_brain(user_query)
    print("\n[Plan]", plan)

    # Step 2: Execute
    result = execute_plan(plan)

    # Step 3: Final Response
    final_output = f"""
Best Cake: {result['best_cake']['name']}
Price: ₹{result['best_cake']['price']}
Shop: {result['best_cake']['shop']}

Recommendation: Best value chocolate cake under your budget 🎂
"""
    return final_output
response = shopping_agent("Find best chocolate cake under 500")
print("\nFinal Answer:\n", response)
```

```python
# Scenario: The University’s AI Plagiarism Agent
# Imagine a university has deployed an AI-powered plagiarism detection agent named PlagiarismLLM.
# Its job is to monitor student essay submissions in real time and decide whether they are original or potentially plagiarized.

# Step 1: Thinking (Reasoning)
# Whenever an essay is submitted, PlagiarismLLM first analyzes the text:
# - Example: “Essay contains large verbatim passages from online sources”
# - PlagiarismLLM thinks: “Analyzing essay: multiple verbatim passages detected.”

# Step 2: Planning (Choosing an Action)
# Based on the thought, PlagiarismLLM decides which tool to use:
# - If the essay has long verbatim matches or suspicious similarity scores, it plans to flag as potential plagiarism.
# - Otherwise, it plans to mark as original work.

# Step 3: Acting (Running Tools)
# PlagiarismLLM hands over the decision to PlagiarismTools, which executes the action:
# - flag_plagiarism → returns “Potential Plagiarism Detected”
# - mark_original → returns “Original Work”

# Step 4: Answering (Final Decision)
# Once a clear classification is reached, PlagiarismLLM announces the decision:
# - “Decision: Potential Plagiarism Detected”
# - “Decision: Original Work”
# 🎓 AI Plagiarism Detection Agent

# Step 1 — Tools
def analyze_text(essay):
    # Simulated analysis
    if "copied" in essay.lower():
        return {"similarity": 0.85, "verbatim": True}
    else:
        return {"similarity": 0.20, "verbatim": False}

def flag_plagiarism():
    return "Potential Plagiarism Detected"

def mark_original():
    return "Original Work"


# Step 2 — Agent Brain (Thinking + Planning)
def agent_brain(analysis):
    plan = []

    print("\n[Thinking] Analyzing essay results...")
    print(f"Similarity Score: {analysis['similarity']}, Verbatim: {analysis['verbatim']}")

    if analysis["similarity"] > 0.7 or analysis["verbatim"]:
        print("[Decision] High similarity detected → Flag plagiarism")
        plan.append("flag_plagiarism")
    else:
        print("[Decision] Low similarity → Mark as original")
        plan.append("mark_original")

    return plan


# Step 3 — Execution Engine (Acting)
def execute_plan(plan):
    context = {}

    for step in plan:
        if step == "flag_plagiarism":
            print("\n[Action] Flagging plagiarism...")
            context["result"] = flag_plagiarism()
            print(context["result"])

        elif step == "mark_original":
            print("\n[Action] Marking as original...")
            context["result"] = mark_original()
            print(context["result"])

    return context


# Step 4 — Full Agent
def plagiarism_agent(essay):
    print("Essay Submitted:", essay)

    # Step 1: Analyze
    analysis = analyze_text(essay)

    # Step 2: Plan
    plan = agent_brain(analysis)
    print("\n[Plan]", plan)

    # Step 3: Execute
    result = execute_plan(plan)

    # Step 4: Final Answer
    final_output = f"""
Decision: {result['result']}
Similarity Score: {analysis['similarity']}
"""
    return final_output


# Step 5 — Run It
response = plagiarism_agent("This essay contains copied content from internet sources")
print("\nFinal Answer:\n", response)
```

```python
import requests

from google.colab import userdata

API_URL = "https://openrouter.ai/api/v1/chat/completions"
api_key = userdata.get('OPENAI_API_KEY')

HEADERS = {
    "Authorization": f"Bearer {api_key}",
    "Content-Type": "application/json"
}


# 🔹 LLM Function (Simple Inference API)
def query_llm(prompt: str):
    payload = {
        "model": "openai/gpt-3.5-turbo",
        "messages": [
            {"role": "user", "content": prompt}
        ]
    }

    try:
        response = requests.post(API_URL, headers=HEADERS, json=payload)
        data = response.json()

        return data["choices"][0]["message"]["content"]

    except Exception as e:
        return f"Error: {str(e)}"


# 🔹 Tools
def flag_suspicious(transaction):
    return "Suspicious Transaction"

def mark_normal(transaction):
    return "Normal Transaction"


# 🔹 Agent
def fraud_agent(transaction: str):
    print("Transaction:", transaction)

    # Step 1: Thinking (LLM reasoning)
    thought = query_llm(f"Analyze this transaction for fraud risk: {transaction}")
    print("\n[Thinking]\n", thought)

    # Step 2: Planning (Rule-based decision)
    if "overseas" in transaction.lower() or "multiple deposits" in transaction.lower():
        print("\n[Plan] Flag as suspicious")
        action = flag_suspicious
    else:
        print("\n[Plan] Mark as normal")
        action = mark_normal

    # Step 3: Acting
    print("\n[Action] Executing...")
    result = action(transaction)
    print(result)

    # Step 4: Final Answer (LLM explanation)
    explanation = query_llm(
        f"Explain why this transaction is classified as: {result}. Transaction: {transaction}"
    )

    return f"""
Final Decision: {result}

Explanation:
{explanation}
"""


# 🔹 Run Example
print(fraud_agent("Payment to overseas account"))
```

```python
# 🏥 Scenario: The Hospital’s AI Appointment Agent
# Imagine a hospital has deployed an AI-powered agent named MediLLM.
# Its job is to monitor patient appointment requests and decide whether the requested slot is available or needs rescheduling.

# Step 1 — Thinking (Reasoning)
# Whenever a patient request comes in, MediLLM first analyzes the details:
# - Example: “Request for Dr. Sharma at 10 AM tomorrow”
# - MediLLM thinks: “Analyzing appointment request: Dr. Sharma, 10 AM tomorrow.”

# Step 2 — Planning (Choosing an Action)
# Based on the thought, MediLLM decides which tool to use:
# - If the requested slot is free, it plans to confirm appointment.
# - If the slot is already booked, it plans to suggest rescheduling.

# Step 3 — Acting (Running Tools)
# MediLLM hands over the decision to AppointmentTools, which executes the action:
# - confirm_slot → returns “Appointment Confirmed”
# - suggest_reschedule → returns “Slot Unavailable, Please Reschedule”

# Step 4 — Answering (Final Decision)
# Once a clear classification is reached, MediLLM announces the decision:
# - “Decision: Appointment Confirmed”
# - “Decision: Slot Unavailable, Please Reschedule”
import requests

# Replace with your Hugging Face token
from google.colab import userdata

API_URL = "https://openrouter.ai/api/v1/chat/completions"
api_key = userdata.get('OPENAI_API_KEY')

HEADERS = {
    "Authorization": f"Bearer {api_key}",
    "Content-Type": "application/json"
}



# 🔹 LLM Function (Thinking)
def query_llm(prompt: str):
    payload = {
        "model": "openai/gpt-3.5-turbo",
        "messages": [
            {"role": "user", "content": prompt}
        ]
    }

    try:
        response = requests.post(API_URL, headers=HEADERS, json=payload)
        data = response.json()

        return data["choices"][0]["message"]["content"]

    except Exception as e:
        return f"Error: {str(e)}"


# 🔹 Tools (Actions)
def confirm_slot(request):
    return "Appointment Confirmed"

def suggest_reschedule(request):
    return "Slot Unavailable, Please Reschedule"


# 🔹 Simulated Database (Booked Slots)
booked_slots = [
    {"doctor": "Dr. Sharma", "time": "10 AM tomorrow"},
    {"doctor": "Dr. Mehta", "time": "2 PM today"}
]


# 🔹 Agent
def medi_agent(request: str):
    print("Patient Request:", request)

    # Step 1: Thinking
    thought = query_llm(f"Analyze this appointment request: {request}")
    print("\n[Thinking]\n", thought)

    # Step 2: Planning (Check availability)
    is_booked = any(
        slot["doctor"].lower() in request.lower() and
        slot["time"].lower() in request.lower()
        for slot in booked_slots
    )

    if is_booked:
        print("\n[Plan] Slot unavailable → Reschedule")
        action = suggest_reschedule
    else:
        print("\n[Plan] Slot available → Confirm")
        action = confirm_slot

    # Step 3: Acting
    print("\n[Action] Executing...")
    result = action(request)
    print(result)

    # Step 4: Final Answer (LLM explanation)
    explanation = query_llm(
        f"Explain this decision: {result} for request: {request}"
    )

    return f"""
Decision: {result}

Explanation:
{explanation}
"""


# 🔹 Run Example
print(medi_agent("Request for Dr. Sharma at 10 AM tomorrow"))
```
