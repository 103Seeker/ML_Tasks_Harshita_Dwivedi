```python
# Scenario: Automated Business Report Assistant
# 🎯 Problem Statement
# In companies:
# - Managers spend hours compiling data from multiple sources (CRM, ERP, spreadsheets).
# - Reports often require manual formatting, chart creation, and consistency checks.
# - This slows down decision-making and introduces human error.
# 👉 Solution
# A multi-agent AutoGen system where:
# - 🤖 Assistant Agent drafts report templates and queries relevant data.
# - ⚙️ UserProxy Agent executes data pulls and formatting automatically.
# - 🔁 The system iterates until the report meets formatting, accuracy, and completeness standards.
# 🔄 Workflow
# Business Request → AI drafts report structure → Data fetched & formatted → Charts auto-generated → Validation loop → Final polished report
import os
import autogen
from google.colab import userdata

#  Load API Key
api_key = userdata.get("api_key")

# Safety check
if not api_key or not api_key.startswith("gsk_"):
    raise ValueError("❌ Invalid Groq API Key")

os.environ["api_key"] = api_key

#  Working directory
os.makedirs("coding", exist_ok=True)

#  LLM Config
config_list = [
    {
        "model": "llama-3.3-70b-versatile",
        "api_key": api_key,
        "base_url": "https://api.groq.com/openai/v1"
    }
]

llm_config = {"config_list": config_list, "temperature": 0}

#  Assistant Agent (Report Generator)
assistant = autogen.AssistantAgent(
    name="Report_Generator",
    llm_config=llm_config,
    system_message="""
You are a business analyst AI.

Your job:
1. Generate structured reports
2. Write Python code to:
   - Analyze data
   - Create charts
3. Explain insights clearly

Always:
- Use pandas & matplotlib
- Print clean outputs
"""
)

#  Reviewer Agent (Validation Loop)
reviewer = autogen.AssistantAgent(
    name="Reviewer",
    llm_config=llm_config,
    system_message="""
You are a strict reviewer.

Check:
- Data correctness
- Completeness
- Business insights quality

If something is missing → ask for improvement.
If good → say 'APPROVED'
"""
)

#  Executor Agent (runs code)
user_proxy = autogen.UserProxyAgent(
    name="Executor",
    human_input_mode="NEVER",
    max_consecutive_auto_reply=10,
    code_execution_config={
        "work_dir": "coding",
        "use_docker": False
    }
)

#  Group Chat (multi-agent system)
groupchat = autogen.GroupChat(
    agents=[assistant, reviewer, user_proxy],
    messages=[],
    max_round=10
)

manager = autogen.GroupChatManager(
    groupchat=groupchat,
    llm_config=llm_config
)

#  BUSINESS TASK
user_proxy.initiate_chat(
    manager,
    message="""
Create a business report for sales data:

Data: [12000, 15000, 10000, 18000, 13000]

Requirements:
1. Summary statistics (mean, median, min, max)
2. Generate a bar chart
3. Identify trends
4. Provide business insights
5. Ensure report is clean and complete

Reviewer must approve the final output.
"""
)
```

```python
# Scenario: AI-Powered HR Onboarding System (Corporate HR Team)
# 🎯 Problem Statement
# In companies:
# - HR teams manually prepare onboarding documents and checklists.
# - Managers validate compliance with policies.
# - Employees often face delays in receiving resources and training schedules.
# 👉 Solution
# A multi-agent system where:
# - 📝 HR Agent prepares onboarding documents, training schedules, and resource lists.
# - ✅ Compliance Agent reviews documents for policy adherence and legal requirements.
# - 📂 Resource Agent allocates accounts, devices, and access permissions.
# - 🔁 The loop continues until all onboarding steps are validated and complete.
# 🔄 Workflow
# Onboarding Request → HR Agent drafts documents → Compliance Agent validates → Resource Agent provisions accounts/devices → Iteration loop → Final approved onboarding package
# 🧹 Conflict Resolution
# - Resolve conflicting policy requirements (e.g., different regional compliance rules).
# - Standardize onboarding templates across departments.
# - Ensure system avoids duplicate resource allocation (e.g., multiple accounts for same employee).
# Install dependencies
# Install dependency
!pip install pyautogen==0.2.25

import os
import time
import autogen
from google.colab import userdata

#  Load API Key
api_key = userdata.get("api_key")

if not api_key:
    raise ValueError(" Missing Groq API Key")

os.environ["api_key"] = api_key

#  LLM Config (LOW COST + FAST)
config_list = [
    {
        "model": "llama-3.1-8b-instant",
        "api_key": api_key,
        "base_url": "https://api.groq.com/openai/v1"
    }
]

llm_config = {
    "config_list": config_list,
    "temperature": 0,
    "cache_seed": 42
}

#  HR Agent
hr_agent = autogen.AssistantAgent(
    name="HR_Agent",
    llm_config=llm_config,
    system_message="""
You are an HR onboarding specialist.

Generate:
- Onboarding document
- 2-week training schedule
- Resource checklist

Keep output concise and structured.
"""
)

#  Compliance Agent
compliance_agent = autogen.AssistantAgent(
    name="Compliance_Agent",
    llm_config=llm_config,
    system_message="""
You are a compliance officer.

Check:
- Legal (based on employee location)
- Company policy adherence

If issues:
→ Return fixes

If correct:
→ Say 'COMPLIANT'
"""
)

#  Resource Agent
resource_agent = autogen.AssistantAgent(
    name="Resource_Agent",
    llm_config=llm_config,
    system_message="""
You handle provisioning.

Allocate:
- Email
- Slack
- GitHub
- Laptop

Rules:
- No duplicate resources
- Role-based access

Return structured list.
"""
)

#  Auditor Agent
auditor = autogen.AssistantAgent(
    name="Auditor",
    llm_config=llm_config,
    system_message="""
Check final output for:

- Duplicate resources
- Missing steps

If OK:
→ 'AUDIT PASSED'

Else:
→ Explain issue
"""
)

# ⚙️ Executor (needed for AutoGen flow)
executor = autogen.UserProxyAgent(
    name="Executor",
    human_input_mode="NEVER",
    code_execution_config=False
)

#  TASK (shortened to save tokens)
task = """
Onboard Software Engineer in Germany.

Generate:
- onboarding doc
- 2-week plan
- resources

Rules:
- Follow EU compliance
- No duplicate resources

Approval needed:
COMPLIANT + AUDIT PASSED
"""

#  RETRY FUNCTION (handles rate limits)
def safe_chat(agent, message):
    while True:
        try:
            return agent.generate_reply(messages=[{"role": "user", "content": message}])
        except Exception as e:
            if "rate_limit" in str(e).lower():
                print("⏳ Rate limit hit. Waiting 60s...")
                time.sleep(60)
            else:
                raise e

#  SEQUENTIAL FLOW (efficient architecture)

print(" HR Agent working...")
hr_output = safe_chat(hr_agent, task)
print(hr_output)

print("\n4 Compliance check...")
compliance_output = safe_chat(compliance_agent, hr_output)
print(compliance_output)

# If not compliant → fix loop
if "COMPLIANT" not in str(compliance_output):
    print("\n🔁 Fixing compliance issues...")
    hr_output = safe_chat(hr_agent, f"Fix issues:\n{compliance_output}")
    compliance_output = safe_chat(compliance_agent, hr_output)

print("\n📂 Resource allocation...")
resource_output = safe_chat(resource_agent, hr_output)
print(resource_output)

print("\n🧾 Audit check...")
final_output = f"{hr_output}\n\n{resource_output}"
audit_output = safe_chat(auditor, final_output)
print(audit_output)

# 🔁 Final validation loop
if "AUDIT PASSED" not in str(audit_output):
    print("\n🔁 Fixing audit issues...")
    resource_output = safe_chat(resource_agent, audit_output)
    final_output = f"{hr_output}\n\n{resource_output}"
    audit_output = safe_chat(auditor, final_output)

print("\n🎉 FINAL OUTPUT:")
print(final_output)
```
